```python
import pandas as pd
import argparse
import logging
from pathlib import Path
import json
```

```python
# Configure logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
```

```python
def clean_data(df: pd.DataFrame, missing: str, date_cols: list, rename_map: dict) -> pd.DataFrame:
    """Clean and normalize data."""
    # Handle missing values (fill with empty string or drop)
    if missing == "drop":
        df = df.dropna()
    elif missing == "zero":
        df = df.fillna(0)
    else: # default: empty string
        df = df.fillna("")

    #parse date columns if present
    for col in date_cols:
      if col in df.columns:
        try:
          df[col] = pd.to_datetime(df[col], errors="coerce")
        except Exception as e:
          logging.error(f"Error parsing date column {col}: {e}")

    #simple column renames
    df = df.rename(columns={k: v for k, v in rename_map.items() if k in df.columns})
    return df
```

```python
def convert_csv_to_excel(input_file: str, output_file: str, missing: str, date_cols: list, rename_map: dict):
    """Convert CSV to Excel and clean data."""
    try:
      logging.info(f"Reading CSV file: {input_file}")
      df = pd.read

      logging.info(f"Cleaning data...")
      df = clean_data(df, missing, date_cols, rename_map)

      logging.info(f"Writing Excel file: {output_file}")
      df.to_excel(output_file, index=False, engine="openpyxl")

      logging.info(f"Successfully converted CSV to Excel: {output_file}")

    except Exception as e:
      logging.error(f"Error reading CSV file: {e}")
```

```python
def main():
    parser = argparse.ArgumentParser(description="CSV → Excel Converter")
    parser.add_argument("-i", "--input", required=True, help="Path to input CSV file")
    parser.add_argument("-o", "--output", required=True, help="Path to output Excel file (.xlsx)")
    args = parser.parse_args()

    input_path = Path(args.input)
    output_path = Path(args.output)

    if not input_path.exists():
        logging.error(f"Input file does not exist: {input_path}")
        return

    convert_csv_to_excel(str(input_path), str(output_path))

if __name__ == "__main__":
    main()
```
