```python
import json
```

```python
class Book:
  def __init__(self, title, author, isbn):
    self.title = title
    self.author = author
    self.isbn = isbn
    self.issued = False

  def issue(self):
    if self.issued:
      self.issued = True
      return True
    return False

  def return_book(self):
    if self.issued:
      self.issued = False
      return True
    return False

  def to_dict(self):
    return {
      'title': self.title,
      'author': self.author,
      'isbn': self.isbn,
      'issued': self.issued
    }

  @staticmethod
  def from_dict(book_dict):
    book = Book(book_dict['title'], book_dict['author'], book_dict['isbn'])
    book.issued = book_dict['issued']
    return book
```

```python
class Library:

  def __init__(self):
    self.books = {}

  def add_book(self, book):
    self.books[book.isbn] = book

  def search_by_title(self, title):
    return [book for book in self.books.values() if title.lower() in book.title.lower()]

  def search_by_author(self, author):
    return [book for book in self.books.values() if author.lower() in book.author.lower()]

  def issue_book(self, isbn):
    if isbn in self.books:
      return self.books[isbn].issue()
    return False

  def return_book(self, isbn):
    if isbn in self.books:
      return self.books[isbn].return_book()
    return False

  def report_total_books(self):
    return len(self.books)

  def report_issued_books(self):
    return sum(1 for book in self.books.values() if book.issued)

  def save_to_file(self, filename="library.json"):
    with open(filename, 'w') as f:
      json.dump([book.to_dict() for book in self.books.values()], f)

  def load_from_file(self, filename="library.json"):
    try:
      with open(filename, 'r') as f:
        book_dicts = json.load(f)
        for book_dict in book_dicts:
          book = Book.from_dict(book_dict)
          self.add_book(book)
    except FileNotFoundError:
      pass
```

```python
def main():
    library = Library()
    library.load_from_file()

    while True:
        print("\n--- Library Menu ---")
        print("1. Add Book")
        print("2. Search by Title")
        print("3. Search by Author")
        print("4. Issue Book")
        print("5. Return Book")
        print("6. Report Total Books")
        print("7. Report Issued Books")
        print("8. Save & Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            title = input("Enter title: ")
            author = input("Enter author: ")
            isbn = input("Enter ISBN: ")
            library.add_book(Book(title, author, isbn))
            print("Book added successfully.")

        elif choice == "2":
            title = input("Enter title to search: ")
            results = library.search_by_title(title)
            for b in results:
                print(f"{b.title} by {b.author} (ISBN: {b.isbn}, Issued: {b.issued})")

        elif choice == "3":
            author = input("Enter author to search: ")
            results = library.search_by_author(author)
            for b in results:
                print(f"{b.title} by {b.author} (ISBN: {b.isbn}, Issued: {b.issued})")

        elif choice == "4":
            isbn = input("Enter ISBN to issue: ")
            if library.issue_book(isbn):
                print("Book issued successfully.")
            else:
                print("Book not available or already issued.")

        elif choice == "5":
            isbn = input("Enter ISBN to return: ")
            if library.return_book(isbn):
                print("Book returned successfully.")
            else:
                print("Book not found or not issued.")

        elif choice == "6":
            print("Total books:", library.report_total_books())

        elif choice == "7":
            print("Issued books:", library.report_issued_books())

        elif choice == "8":
            library.save_to_file()
            print("Library saved. Exiting...")
            break

        else:
            print("Invalid choice. Try again.")


if __name__ == "__main__":
    main()
```
