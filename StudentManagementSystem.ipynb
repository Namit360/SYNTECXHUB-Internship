```python
import json
import os
```

```python
class Student:
  def __init__(self, studentid, name, grade):
    self.studentid = studentid
    self.name = name
    self.grade = grade

  def __str__(self):
    return f"ID : {self.studentid:<5} Name : {self.name:<15} Grade : {self.grade:<3}"

```

## StudentManager class responsible for managing student records. This class will hold the collection of students and encapsulate the operations.

```python
class StudentManager:
  def __init__(self, filename="students.json"):
    self.filename = filename
    self.students = []
    self.load_students()

  def load_students(self):
    # os.path.exists(self.filename) verifies whether the file (whose path is
    # stored in self.filename) actually exists
    if os.path.exists(self.filename):
      #Opens the file safely using a with statement (ensures the file is closed automatically after use).
      #"r" means read-only mode.
      with open(self.filename, "r") as file:
        data = json.load(file)
        #Iterates over each dictionary student_data in data.
        #Student(**student_data) uses dictionary unpacking to pass the dictionary keys/values as arguments to the Student class constructor.
        self.students = [Student(**student_data) for student_data in data]

  def save_students(self):
    with open(self.filename, "w") as file:
      #data = [student.__dict__ for student in self.students]
      json.dump([student_data.__dict__ for student_data in self.students], file)

  def add_student(self, studentid, name, grade):
    if any(student.studentid == studentid for student in self.students):
      print("Student already exists")
      return
    self.students.append(Student(studentid, name, grade))
    self.save_students()
    print("Stuent added successfully")

  def update_student(self, studentid, name, grade):
    for student in self.students:
      if student.studentid == studentid:
        if name:
          student.name = name
        if grade:
          student.grade = grade
        self.save_students()
        print("Student updated successfully")
        return
    print("Student not found")

  def delete_student(self, studentid):
    for student in self.students:
      if student.studentid == studentid:
        self.students.remove(student)
        self.save_students()
        print("Student deleted successfully")
        return
    print("Student not found")

  def list_students(self):
    if not self.students:
      print("student list is empty")
      return
    else:
      print("\n --- Students Records ---")
      for student in self.students:
        print(student)
```

```python
def main():
  student_manager = StudentManager()

  while True:
    print("\n --- Student Management System --- ")
    print("\n 1. Add Student")
    print("\n 2. Update Student")
    print("\n 3. Delete Student")
    print("\n 4. List Students")
    print("\n 5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
            studentid = input("Enter ID: ")
            name = input("Enter Name: ")
            grade = input("Enter Grade: ")
            student_manager.add_student(studentid, name, grade)

    elif choice == "2":
            studentid = input("Enter ID to update: ")
            name = input("Enter new Name (leave blank to skip): ")
            grade = input("Enter new Grade (leave blank to skip): ")
            student_manager.update_student(studentid, name if name else None, grade if grade else None)

    elif choice == "3":
            studentid = input("Enter ID to delete: ")
            student_manager.delete_student(studentid)

    elif choice == "4":
            student_manager.list_students()

    elif choice == "5":
            print("Exiting...")
            break

    else:
            print("Invalid choice. Try again.")
```

```python
if __name__ == "__main__":
    main()
```
